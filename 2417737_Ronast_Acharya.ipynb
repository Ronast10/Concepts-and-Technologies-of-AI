Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

```python
# Calculate Q1, Q3, and IQR for Score
score_Q1 = south_asia_data['score'].quantile(0.25)
score_Q3 = south_asia_data['score'].quantile(0.75)
score_IQR = score_Q3 - score_Q1

# Calculate Q1, Q3, and IQR for GDP per Capita
gdp_Q1 = south_asia_data['Log GDP per capita'].quantile(0.25)
gdp_Q3 = south_asia_data['Log GDP per capita'].quantile(0.75)
gdp_IQR = gdp_Q3 - gdp_Q1

# Define outlier bounds
score_lower_bound = score_Q1 - 1.5 * score_IQR
score_upper_bound = score_Q3 + 1.5 * score_IQR

gdp_lower_bound = gdp_Q1 - 1.5 * gdp_IQR
gdp_upper_bound = gdp_Q3 + 1.5 * gdp_IQR

# Identify outliers
outliers = south_asia_data[
    (south_asia_data['score'] < score_lower_bound) | (south_asia_data['score'] > score_upper_bound) |
    (south_asia_data['Log GDP per capita'] < gdp_lower_bound) | (south_asia_data['Log GDP per capita'] > gdp_upper_bound)
]

# Display outliers
outliers
```

2. Define outliers using the 1.5 × IQR rule.


The Interquartile Range (IQR) is the range between the 25th percentile (Q1) and the 75th percentile (Q3). Outliers are defined as values that fall below:

Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR

This is the formula we use to detect outliers for Score and GDP per Capita.

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.


```python
# Scatter plot with outliers highlighted
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Log GDP per capita',
    y='score',
    data=south_asia_data,
    label='Non-Outliers',
    color='black'
)
sns.scatterplot(
    x='Log GDP per capita',
    y='score',
    data=outliers,
    label='Outliers',
    color='blue'
)
plt.title('Outlier Detection: GDP per Capita vs. Score')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.legend()
plt.show()
```

4. Discuss the characteristics of these outliers and their potential impact on regional averages.

Outliers are countries that differ significantly from the rest, either due to an exceptionally high GDP per capita or an unusually high or low score. These outliers can distort the overall averages for the entire region.

When a country is much wealthier than the others, it can inflate the region's wealth, making it appear richer than it actually is. Conversely, a significantly poorer country can create the impression that the region is in worse economic shape than it truly is.

Outliers can also exaggerate the range of the data, giving the impression that the countries within the region are more diverse or different from one another than they actually are.
